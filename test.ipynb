```python
import pandas as pd

df = pd.read_excel("equipment.xlsx")
df = df.rename(columns={
    "№": "Number",
    "Раздел": "Section",
    "Помещение": "Room",
    "Наименование_2025": "Name",
    "Ед. изм.": "Unit",
    "КОЛ-ВО на кабинет": "Count",
})
rooms = {'universal': 22, 'foreign_lang': 4, 'biology_safety': 2, 'physics': 3, 'chemistry': 2, 'it_info': 1, 'it_material': 0, 'informatics': 2}   

subject_name_map = {
        'universal': ['универсальный учебный кабинет', 'учебные кабинеты (1-е классы)', 'учебные кабинеты (2-4 классы)'],
        'foreign_lang': ['кабинет иностранного языка'],
        'biology_safety': ['биология/обзр', 'основ безопасности и защиты родины'], # Специальный случай
        'physics': ['физика'],
        'chemistry': ['химия'],
        'it_info': ['информатика', 'информационных технологий'],
        'it_material': ['материальной технологии', 'творческая мастерская'], # Специальный случай
        'informatics': ['информатика', 'информационных технологий']
    }

df["Room"].unique()
```

```python
room_multiplier_map = {}

for key, names_list in subject_name_map.items():
    qty = rooms.get(key, 0) # Получаем количество кабинетов
    for name in names_list:
        room_multiplier_map[name.lower().strip()] = qty

df['Multiplier'] = df['Room'].astype(str).str.lower().str.strip().map(room_multiplier_map).fillna(0)
df['Total_Count'] = df['Count'] * df['Multiplier']
result_df = df[['Name', 'Total_Count', 'Unit']].rename(columns={'Total_Count': 'Count'})

```

```python
result_df = result_df[result_df["Count"] > 0]
result_df
```
